```python
a.Gender = a.Gender.fillna("Male") 
a.Married = a.Married.fillna("Yes")
a.Self_Employed = a.Self_Employed.fillna("No")
a.Credit_History = a.Credit_History.fillna(0.0)
```

```python
a.Dependents = a.Dependents.fillna(a.Dependents.mean())
a.LoanAmount = a.LoanAmount.fillna(a.LoanAmount.mean())
a.Loan_Amount_Term = a.Loan_Amount_Term.fillna(a.Loan_Amount_Term.median())

```

```python
a.isnull().sum()
```

```python
# a.Gender = a.Gender.fillna("Male")  # correct
a.Gender.fillna("Male" , inplace = True)  # correct 
# a.Gender = a.Gender.fillna("Male" , inplace = True)  # wrong
```

```python
a.Loan_Status.replace({'Y':1 , 'N':0 }, inplace = True)
a.Education.replace({'Graduate':1 , 'Not Graduate':0 }, inplace = True)
a.Self_Employed.replace({'Yes':1 , 'No':0 }, inplace = True)
a.Property_Area.replace({'Semiurban':1,'Urban':2 , 'Rural':0 }, inplace = True)
a.Married.replace({'Yes':1 , 'No':0 }, inplace = True)
a.Gender.replace({'Male':1 , 'Female':0 }, inplace = True)


```
